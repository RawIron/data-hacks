## plot univariate functions

### subplot

compare log, sqrt and linear
use subplot to zoom in on two intervals of the definition 

```python
ax = plt.subplot(1,2,1)
plt.axis([-1, 10, -1, 5])
plt.grid(True)

d = linspace(0.1, 10, 100)

plot(d, sqrt(d), label='sqrt')
plot(d, log(d) + np.e, label='log')
plot(d, d, label='id')

plt.legend(loc='lower right')


ax = plt.subplot(1,2,2)
plt.axis([-0.1, 1, -1, 3])
plt.grid(True)

d = linspace(0.01, 1, 100)

plot(d, sqrt(d), label='sqrt')
plot(d, log(d) + np.e, label='log')
plot(d, d + 0.25, label='id')

plt.legend(loc='lower right')
```

### 1 over x

```python
plt.axis([-1, 10, -1, 5])
plt.grid(True)

d = linspace(0.1, 10, 100)

plot(d, 1/d, label='1/x')

plt.legend(loc='lower right')
```

```python
plt.axis([-5, 5, -10, 1])
plt.grid(True)

d = linspace(-10, 10, 100)
func = np.vectorize(lambda x: x * (1 - x))

plot(d, func(d))
```

## plot data

### univariate data

```python
data = np.random.randn(1000)
print data.mean()
print data.std()

plt.grid(True)
n, bins, patches = plt.hist(data, 30, normed=1, facecolor='g', alpha=.5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(0, len(data)), data, '.')
plt.show()
```

### multivariate data

```python
plt.scatter(np.arange(0, len(data)), data, s=2)
plt.show()

plt.scatter(data, data, s=2)
plt.show()
```

```python
x = np.random.normal(12, 4, 200)
y = np.random.normal(2, 1, 200)

plt.scatter(x, y, s=2)
plt.show()
```
